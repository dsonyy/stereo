```python
import stereo as st
import numpy as np
import cv2 as cv
import PIL.Image
```

```python
calibration = st.Calibration2Cams()
_ = calibration.load("calibrated/2cam-csi-12cm-v1.yaml")
```

```python
cameras = st.Cameras([0, 1], transformations=[
    (cv.cvtColor, cv.COLOR_BGR2GRAY)
])
cameras.to_csi_device()
```

```python
stereo = st.StereoVision2Cams(calibration)
```

```python
cameras.capture()
depth = stereo.calculate_depth(cameras.frames[0], cameras.frames[1])
depth2 = ((depth - depth.min()) * 255).astype(np.uint8)
cameras.nshow()
PIL.Image.fromarray(depth2)
```
